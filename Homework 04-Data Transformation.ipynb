#### **Q7** How many of the origin airport codes in the flights dataset do not have a mapping to a real airport in the aiports file?  For example, some values of ORIGIN_AIRPORT are numeric, eg 14869.  These appear to be not valid airport codes per the aiports file.)
#### (Hint: There are many ways to solve this one.  Here is one approach: Left join the flights data (flights.csv) with the airports data (airports.csv) using the column ORIGIN_AIRPORT.  Then count the nulls in any of the new fields that are coming from the 'right' table, eg AIRPORT.)

```python
# We can do the merge as described in the question, and count the sum of the isnull rows
df.merge(df_airports, how = 'left' , left_on = 'ORIGIN_AIRPORT', right_on = 'IATA_CODE').AIRPORT.isnull().sum()
```

```python
# Or we can just do the merge and look for the number of rows at the bottom
df[df.merge(df_airports, how = 'left' , left_on = 'ORIGIN_AIRPORT', right_on = 'IATA_CODE').AIRPORT.isnull()]
```
