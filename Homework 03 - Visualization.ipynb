Q1 - What chart type is best suited to show comparisons across categories?

Line Chart  
Bar Chart  
Histogram  
Scatterplot  

Answer:  
Bar charts are used to show categorical data.  Line charts, histograms and scatterplots are better suited to time series, continuous variables, and to show correlation between two variables respectively.

**Q6**  
What chart type is commonly used to display the distribution of a single continuous variable?

- Scatterplot
- Boxplot
- Pie chart
- Histogram

Answer: A single continuous variable is best displayed using a histogram.  Pie-charts and scatterplots require multiple variables.  A box plot can also be used for understanding the distribution of a single continuous variable, but a histogram provides a more complete view.

**Q10**  
Which type of chart is most suitable for displaying frequency distributions?

Answer: Frequency distributions are best represented using histograms.  Scatterplots display relationships, line charts help show time-series data, and tree maps show portions of a whole with rectangles.

**Q11**  
What is the key advantage of using a logarithmic scale in a chart?

Answer: It compresses data with large ranges.

Due to the way the log function behaves, applying the log has the effect of suppressing large numbers while increasing the smaller ones.  In a way, it reduces the variance, and in some cases may allow us to look at the data better.

It does not emphasize extreme values, in fact it does quite the opposite of that.  It does not simplify the interpretation of a chart, is not a more accurate representation of the data.
